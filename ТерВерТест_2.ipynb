##Задача 1
Преподаватель выставляет студетам магистратуры "Инжененрия данных" оценки за сданное дз. C равной вероятностью он может поставить любую оценку от 5 до 10. В одной из групп дз сдали 20 человек.
Найдите вероятность того, что ровно четверть студентов получит 10.  

В ответ напишите округленное значение в виде десятичной дроби с двумя цифрами после запятой.


```python
import math
p = 1/6
n = 20
k = 5

probability = math.comb(n, k) * (p**k) * ((1 - p)**(n - k))
print(f'Final probability: {round(probability,2)}')


```

## Задача 2
Преподаватель выставляет студетам магистратуры "Инжененрия данных" оценки за сданное дз. C равной вероятностью он может поставить любую оценку от 5 до 10. В одной из групп дз сдали 20 человек.
Найдите вероятность того, что преподаватель выставит не менее чем пять отличных оценок (***отличные оценки  - это 8,9 или 10***).

В ответ напишите округленное значение в виде десятичной дроби с двумя цифрами после запятой.

```python
import math
p = 3/6
n = 20
probability = 0

for index, k in enumerate(range(5, 21)):
    prob_i = math.comb(n, k) * (p**k) * ((1 - p)**(n - k))
    probability += prob_i
print(f'Final probability: {probability}')


```

## Задача 6
Студент магистратуры "Инженерия данных" решает задачи. Количество задач, которые он решит за час распредлено по закону Пуассона с матожиданием 2. Найдите вероятность того, что он решит больше двух задач за этот час.

В ответ напишите округленное значение в виде десятичной дроби с двумя цифрами после запятой.

```python
import math

Ex = 2

def pois(Ex,k):
  return Ex**k * math.e**(-Ex) / math.factorial(k)

probability = 1 - pois(Ex,0) - pois(Ex,1) - pois(Ex,2)
print(f'Final probability: {round(probability,2)}')
```

##  Задача 7

В поселке 1000 домов, каждый из которых застрахован от пожара в одной страховой компании на сумму 100.000 рублей. Страховой взнос составляет 300 рублей в год. Для данного поселка вероятность пожара в доме в течении года равна 0.002. Какова вероятность того, что в течение года страховая компания по данному виду полиса выплатит больше, чем соберет?

При решении используйте подходящую предельную теорему. В ответ напишите округленное значение в виде десятичной дроби с двумя цифрами после запятой.

```python
import math
n = 1000
p = 0.002
Ex = n * p

def pois(Ex,k):
  return Ex**k * math.e**(-Ex) / math.factorial(k)

probability = 1 - pois(Ex,0) - pois(Ex,1) - pois(Ex,2) - pois(Ex,3)
print(f'Final probability: {round(probability,2)}')
```
